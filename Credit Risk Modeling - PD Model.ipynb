# Computing weigh of evidence and information gain (data preparation)

```python
# Copy data into 2 new dataframes
df_inputs_prepr = loan_data_inputs
df_targets_prepr = loan_data_targets
```

```python
# We are going to compute weight of evidence and info for "grade" feature
df_inputs_prepr['grade'].unique()
```

```python
#create a new dataframe containing "grade" and traget column
df1 = pd.concat([df_inputs_prepr['grade'], df_targets_prepr], axis = 1)
df1.head(5)
```

```python
# Groups the data according to a criterion contained in one column.
# Does not turn the names of the values of the criterion as indexes.
# Aggregates the data in another column, using a selected function.
# In this specific case, we group by the column with index 0 and we aggregate the values of the column with index 1.
# More specifically, we count them.
# In other words, we count the values in the column with index 1 for each value of the column with index 0.
df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].count()
```

```python
# Groups the data according to a criterion contained in one column.
# Does not turn the names of the values of the criterion as indexes.
# Aggregates the data in another column, using a selected function.
# Here we calculate the mean of the values in the column with index 1 for each value of the column with index 0.

df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].mean()
```

```python
# Put mean and count into the same dataframe
df1 = pd.concat([df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].count(),
                df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].mean()], axis = 1)
```

```python
df1
```

```python
# use iloc and fancy indexing to select only required columns

df1 = df1.iloc[:, [0, 1, 3]]
df1
```

```python
# Changes the names of the columns of a dataframe.
df1.columns = [df1.columns.values[0], 'n_obs', 'prop_good']
df1
```

```python
# Compute proportion of total samples

df1['prop_n_obs'] = df1['n_obs'] / df1['n_obs'].sum()
df1
```

```python
# compute the number of good and bad saamples for each grade

df1['n_good'] = df1['prop_good'] * df1['n_obs']
df1['n_bad'] = (1 - df1['prop_good']) * df1['n_obs']
df1
```

```python
# Sanity check
(57431.0 + 2328.0) == 59759
# A better sanity check

df1['n_obs'] == df1['n_good'] + df1['n_bad']
```

```python
# Compute proportion of good and bad loans 
df1['prop_n_good'] = df1['n_good'] / df1['n_good'].sum()
df1['prop_n_bad'] = df1['n_bad'] / df1['n_bad'].sum()
df1
```

```python
# Compute Woe = ln (%good/%bad)

df1['WoE'] = np.log(df1['prop_n_good'] / df1['prop_n_bad'])
```

```python
# Compute information value
df1['IV'] = (df1['prop_n_good'] - df1['prop_n_bad']) * df1['WoE']
df1['IV'] = df1['IV'].sum()
# We sum all values of a given column.
df1
```

```python
# WoE function for discrete unordered variables
# Here we combine all of the operations above in a function.
# The function takes 3 arguments: a dataframe, a string, and a dataframe. The function returns a dataframe as a result.

def woe_discrete(df, discrete_variabe_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variabe_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                    df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    #df = df.sort_values(['WoE'])
    #df = df.reset_index(drop = True)
    #df['diff_prop_good'] = df['prop_good'].diff().abs()
    #df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```

# Handling continuous features

By law, most countries require use of easy to interpret PD model

Because of this, just like we used dummy variables for home (own/rent/mortgage) we will use dummy variable for 

continous features as well

We will divide a contionous feature such as income into bins (categories) and then apply the same process as we did 

before for discrete variables

One different is that with continous variables the order has meaning. 

For example, for contionous variable like income, if we create bins such as 0-10, 10-20, 30-40 and so on. 

The group 50-60 represents a lower income group whereas the group 90-100 presents a higher income group.

Compare this with a discrete variable such as purpose_of_loan. A education loan and a car loan have different purposes 

but they done represent any ordering. 


Note: Read nominal vs ordinal from basic statistics for futher clarification



```python
# WoE function for ordered discrete and continuous variables
# Same function as before difference: we order the results by the values of a different column.
# The function takes 3 arguments: a dataframe, a string, and a dataframe. The function returns a dataframe as a result.
def woe_ordered_continuous(df, discrete_variabe_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variabe_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                    df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    #df = df.sort_values(['WoE']) # commented out because we want to preserve natural ordering
    #df = df.reset_index(drop = True) # not neededss
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```

The next step in credit risk modeling is to compute the p-value for each of the independent varibles.

Recall from statistics class a independent variable is considered statistically significant for predicting the dependent variable if the p-value of the independent variable is less than 0.05.

Note: There is no scientific reason behind picking 0.05 as threshold of p-value. It is somewhat arbitrary.

Sklearn does not have a function to compute p-values.

Therefore we will write the function to compute p-values

We will enhance the inbuilt LogisticRegression class to compute p-values


```python
# P-values for sklearn logistic regression.

# Class to display p-values for logistic regression in sklearn.

from sklearn import linear_model
import scipy.stats as stat

class LogisticRegression_with_p_values (LogisticRegression):
    
    def __init__(self,*args,**kwargs):#,**kwargs):
        self.model = linear_model.LogisticRegression(*args,**kwargs)#,**args)

    def fit(self,X,y):
        self.X = X
        self.model.fit(X,y)
        self.coef_ = self.model.coef_
        self.intercept_ = self.model.intercept_
    
    def return_pvalues(self):
         #### Compute p-values for the fitted model ####
        denom = (2.0 * (1.0 + np.cosh(self.model.decision_function(self.X))))
        denom = np.tile(denom,(self.X.shape[1],1)).T
        F_ij = np.dot((self.X / denom).T,self.X) ## Fisher Information Matrix
        Cramer_Rao = np.linalg.inv(F_ij) ## Inverse Information Matrix
        sigma_estimates = np.sqrt(np.diagonal(Cramer_Rao))
        z_scores = self.coef_[0] / sigma_estimates # z-score for eaach model coefficient
        p_values = [stat.norm.sf(abs(x)) * 2 for x in z_scores] ### two tailed test for p-values
        
        self.p_values = p_values
        return self.p_values
```

 Let us  define a function called 'n_approved' which assigns a value of 1 if a predicted probability
 
is greater than the parameter p, which is a threshold, and a value of 0, if it is not.

 Then it sums the column.
 
 Thus, if given any percentage values, the function will return 
 
the number of rows wih estimated probabilites greater than the threshold. 

```python
#define the function described above
def n_approved(p):
    return np.where(df_actual_predicted_probs['y_hat_test_proba'] >= p, 1, 0).sum()
```

We will treat each sample in the test set as a new application for credit

For each threshold in df_cutoff we will compute the number of applications that will be approved


```python
# Compute number of applications that will be approved

df_cutoffs['N Approved'] = df_cutoffs['thresholds'].apply(n_approved)
```

```python
# Compute number of applications that will be rejected which is total - Number approved

df_cutoffs['N Rejected'] = df_actual_predicted_probs['y_hat_test_proba'].shape[0] - df_cutoffs['N Approved']
```

```python
# Compute approval and reject rates . 

df_cutoffs['Approval Rate'] = df_cutoffs['N Approved'] / df_actual_predicted_probs['y_hat_test_proba'].shape[0]

df_cutoffs['Rejection Rate'] = 1 - df_cutoffs['Approval Rate']
```
